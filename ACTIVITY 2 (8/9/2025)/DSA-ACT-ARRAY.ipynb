```python
def current_array(arr):
  print("----------------------------------")
  print("Current list:")
  for i, element in enumerate(arr):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def traverse_array(arr):
  """Function to traverse and print all elements in the array"""
  current_array(arr)
  print("Traversing the array:")
  for i, element in enumerate(arr):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def reverse(arr):
  current_array(arr)
  print("Reversing all elements:")
  for i, element in enumerate(arr[::-1]):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def sort_descend(arr):
  current_array(arr)
  print("Sorting (Descending) all elements:")
  arr.sort(reverse = True)
  for i, element in enumerate(arr):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def sort_ascend(arr):
  current_array(arr)
  print("Sorting (Ascending) all elements:")
  arr.sort(reverse = False)
  for i, element in enumerate(arr):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def insert(arr):
  current_array(arr)
  print("Inserting an element:")
  index_array = int(input("Enter the index number: "))
  insert_array = int(input("Enter a number to insert: "))
  arr.insert(index_array,insert_array)
  for i, element in enumerate(arr):
      print(f"Index {i}: {element}")
  print("----------------------------------")

def delete(arr):
  current_array(arr)
  print("Deleting an element:")
  delete_index = int(input("Enter index number to remove: "))
  arr.pop(delete_index)
  print("Updated list:")
  for i, element in enumerate(arr):
    print(f"Index {i}: {element}")
  print("----------------------------------")

def modify(arr):
  current_array(arr)
  print("Modifying an element:")
  index_array = int(input("Enter the index number to modify: "))
  replace_element = int(input("Enter a number: "))
  arr[index_array] = replace_element
  print("Updated list:")
  for i, element in enumerate(arr):
    print(f"Index {i}: {element}")
  print("----------------------------------")

def main ():
  arr = [10, 100, 0, -100, -10]
  print("Original array:", arr)

  while True:
    print("\nMenu:")
    print("1. Traverse array")
    print("2. Reverse array")
    print("3. Sort ascending")
    print("4. Sort descending")
    print("5. Insert")
    print("6. Delete")
    print("7. Modify")
    print("8. Exit")
    choice = input("Enter your choice (1-8): ")

    if choice == "1":
      traverse_array(arr)
    elif choice == "2":
      reverse(arr)
    elif choice == "3":
      sort_ascend(arr)
    elif choice == "4":
      sort_descend(arr)
    elif choice == "5":
      insert(arr)
    elif choice == "6":
      delete(arr)
    elif choice == "7":
      modify(arr)
    elif choice == "8":
      break
    else:
      print("Invalid choice. Please enter a number between 1-8.")

if __name__ == "__main__":
  main()
```
